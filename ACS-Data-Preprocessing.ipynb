### Create a normalized data set

- Set yes/no to binary 1/0 
- Set numeric to range [0 1]
- Convert categorical to one-hot encoded binary columns, i.e., generate a new binary column for each categorical value

```python
# features/variables
X = []

dft = pd.DataFrame()

dft = df[['CASEID']].astype('int')
X.append('CASEID')

dft['SEX'] = np.where(df[['SEX']]=='male', 1, 0).astype('int') 
X.append('SEX')

CPT_dummies = pd.get_dummies(df.CPT, "CPT")
dft = pd.concat( [dft, CPT_dummies], axis=1) 
#print(CPT_dummies.columns.values)
X = X + list(CPT_dummies.columns.values)

dft['WORKRVU'] = preprocessing.MinMaxScaler().fit_transform(df[['WORKRVU']]).astype('int')
X.append('WORKRVU')

dft['AGE'] = np.where( df['AGE']=='90+', 90, df['AGE'] ).astype('int')
dft['AGE'] = preprocessing.MinMaxScaler().fit_transform(dft[['AGE']]) 
X.append('AGE')

dft['ELECTSURG'] = np.where(df[['ELECTSURG']]=='Yes', 1, 0)
X.append('ELECTSURG')

# Convert height & weight to BMI; [0 1] normalization
#dft['HEIGHT'] = df['HEIGHT']
#dft['WEIGHT'] = df['WEIGHT']
def bmi(row): return(0.453592*row['WEIGHT'])/(0.0254*row['HEIGHT'])
dft['BMI'] = df.apply(lambda row: bmi(row),axis=1)
dft['BMI'] = preprocessing.MinMaxScaler().fit_transform(dft[['BMI']]) 
X.append('BMI')

DIABETES_dummies = pd.get_dummies(df.DIABETES, "DIABETES")
dft = pd.concat( [dft, DIABETES_dummies], axis=1) 
X = X + list(DIABETES_dummies.columns.values)

dft['SMOKE'] = np.where(df[['SMOKE']]=='Yes', 1, 0)
X.append('SMOKE')

DYSPNEA_dummies = pd.get_dummies(df.DYSPNEA, "DYSPNEA")
dft = pd.concat( [dft, DYSPNEA_dummies], axis=1) 
X = X + list(DYSPNEA_dummies.columns.values)

FNSTATUS2_dummies = pd.get_dummies(df.FNSTATUS2, "FNSTATUS2")
dft = pd.concat( [dft, FNSTATUS2_dummies], axis=1) 
X = X + list(FNSTATUS2_dummies.columns.values)

dft['VENTILAT'] = np.where(df[['VENTILAT']]=='Yes', 1, 0)
X.append('VENTILAT')

dft['HXCOPD'] = np.where(df[['HXCOPD']]=='Yes', 1, 0)
X.append('HXCOPD')

dft['ASCITES'] = np.where(df[['ASCITES']]=='Yes', 1, 0)
X.append('ASCITES')

dft['HXCHF'] = np.where(df[['HXCHF']]=='Yes', 1, 0)
X.append('HXCHF')

dft['HYPERMED'] = np.where(df[['HYPERMED']]=='Yes', 1, 0)
X.append('HYPERMED')

dft['RENAFAIL'] = np.where(df[['RENAFAIL']]=='Yes', 1, 0)
X.append('RENAFAIL')

dft['DIALYSIS'] = np.where(df[['DIALYSIS']]=='Yes', 1, 0)
X.append('DIALYSIS')

dft['DISCANCR'] = np.where(df[['DISCANCR']]=='Yes', 1, 0)
X.append('DISCANCR')

dft['STEROID'] = np.where(df[['STEROID']]=='Yes', 1, 0)
X.append('STEROID')

dft['PRSEPIS'] = np.where(df[['PRSEPIS']]=='Yes', 1, 0)
X.append('PRSEPIS')

dft['EMERGNCY'] = np.where(df[['EMERGNCY']]=='Yes', 1, 0)
X.append('EMERGNCY')

ASACLAS_dummies = pd.get_dummies(df.ASACLAS, "ASACLAS")
dft = pd.concat( [dft, ASACLAS_dummies], axis=1) 
X = X + list(ASACLAS_dummies.columns.values)

# target variables

Y = []
# dft['STILLINHOSP'] = np.where(df[['STILLINHOSP']]=='Yes', 1, 0) # > 30 DAYS
# Y.append('STILLINHOSP')

# dft['SEPSHOCKPATOS'] = np.where(df[['SEPSHOCKPATOS']]=='Yes', 1, 0) # > 30 DAYS
# Y.append('SEPSHOCKPATOS')

# dft['REOPERATION1'] = np.where(df[['REOPERATION1']]=='Yes', 1, 0)
# Y.append('REOPERATION1')

# dft['READMISSION1'] = np.where(df[['READMISSION1']]=='Yes', 1, 0)
# Y.append('READMISSION1')

# Predicton dischdest == Expired, add to dataframe for plots and analysis
# DISCHDEST_dummies = pd.get_dummies(df.DISCHDEST, "DISCHDEST")
# dft = pd.concat( [dft, DISCHDEST_dummies], axis=1) 
# dft['DISCHDEST'] = df[['DISCHDEST']]

dft['DISCHDEST_EXPIRED'] = np.where(df[['DISCHDEST']]=='Expired', 1, 0)
Y.append('DISCHDEST_EXPIRED')

dft.drop('CASEID', axis=1, inplace=True)

dft.head(15)
```

### Preliminary data analysis


#### Pain chi-squared goodness-of-fit test
    
Tests whether the distribution of sample categorical data matches an expected distribution.
Chi-squared tests are based on the so-called chi-squared statistic. You calculate the chi-squared statistic with the following formula: 

$sum\Big(\dfrac{(observedâˆ’expected)^2}{expected}\Big)$  


In the formula, observed is the actual observed count for each category and expected is the expected count based on the distribution of the population for the corresponding category. We can use this to see if the value of any factor (categorical) attribute affects the distribution of another attribute. For the purposes of this analysis we will assume the "population" is represented by the entire survey sample.

#### Logistic Regression

Binary Logistic Regression - binary response variable is related to a set of explanatory variables, which can be discrete and/or continuous.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import svm
# X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0

# fit a logistic regression model to the data
model = LogisticRegression()
model.fit(dft[X], dft[Y])
print(model)

# make predictions
expected = dft[Y]
predicted = model.predict(dft[X])

# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))

scores = cross_val_score(model, dft[X], dft[Y]["DISCHDEST_EXPIRED"], cv=10)
print( "Cross validation scores using all features")
print(scores )                                             

# The mean score and the 95% confidence interval of the score estimate are hence given by:
print("Mean accuracy using all features: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

#### SKIP - Support Vector Machine

Discriminative classifier formally defined by a separating hyperplane. In other words, given labeled training data (supervised learning), the algorithm outputs an optimal hyperplane which categorizes new examples.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import svm
# X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0

# fit Support Vector Machine
model = svm.SVC(kernel='linear', C=1)
model.fit(dft[X], dft[Y])
print(model)

# make predictions
expected = dft[Y]
predicted = model.predict(dft[X])

# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))

scores = cross_val_score(model, dft[X], dft[Y]["DISCHDEST_EXPIRED"], cv=10)
print( "Cross validation scores using all features")
print(scores )                                             

# The mean score and the 95% confidence interval of the score estimate are hence given by:
print("Mean accuracy using all features: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

#### Recursive feature elimination to select attributes

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), recursive feature elimination (RFE) selects features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and weights are assigned to each one of them. Then, features whose absolute weights are the smallest are pruned from the current set of features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

Select the 10 most selective attributes
